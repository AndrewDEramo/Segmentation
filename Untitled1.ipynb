```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img=cv2.imread('C:/Users/Andrew/Pictures/parkingLot.jpg')
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
twoDimage=img.reshape((-1,3))
twoDimage=np.float32(twoDimage)
```

```python
plt.imshow(twoDimage)
```

```python
criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

ret,label,center=cv2.kmeans(twoDimage,3,None,criteria,10,cv2.KMEANS_PP_CENTERS)
center=np.uint8(center)
res=center[label.flatten()]
kSegImage3=res.reshape((img.shape))
plt.imshow(kSegImage3)
```

```python
ret,label,center=cv2.kmeans(twoDimage,4,None,criteria,10,cv2.KMEANS_PP_CENTERS)
center=np.uint8(center)
res=center[label.flatten()]
kSegImage4=res.reshape((img.shape))
plt.imshow(kSegImage4)
```

```python
ret,label,center=cv2.kmeans(twoDimage,5,None,criteria,10,cv2.KMEANS_PP_CENTERS)
center=np.uint8(center)
res=center[label.flatten()]
kSegImage5=res.reshape((img.shape))
plt.imshow(kSegImage5)
```

```python
mask=np.copy(kSegImage5)
mask=mask.reshape((-1, 3))
clusters=2
mask[label==clusters] = [0,0,0]

mask=mask.reshape(img.shape)
plt.imshow(mask)
```
